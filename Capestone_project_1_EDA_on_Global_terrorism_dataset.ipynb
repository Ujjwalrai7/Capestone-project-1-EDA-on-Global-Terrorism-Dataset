# **Project Summary -**

The Global Terrorism Database (GTD) is a rich source of information for studying patterns and trends in terrorism over the years. In this project, I performed an exploratory data analysis (EDA) on the GTD to gain insights into the dynamics of terrorism worldwide.

The GTD contains information on over 180,000 terrorist attacks from 1970 to 2017, including the date, location, target, weapon used, and the number of fatalities and injuries. My objective was to explore the data and identify patterns and trends in terrorism over the years.

The first step in my analysis was to load and clean the data. I used the Pandas library in Python to load the dataset and identified missing values using the .isnull() and .sum() methods. I dropped the columns with a high percentage of missing values.

After cleaning the data, I started my analysis by examining the number of terrorist attacks over the years. I created a chart using the Matplotlib library to visualize the trend. The chart showed that the number of terrorist attacks has increased over the years, with a sharp increase after 2000.

Next, I looked at the distribution of terrorist attacks by country. I created a bar chart to visualize the top 10 countries with the most terrorist attacks. The chart showed that Iraq, Pakistan, and Afghanistan have the highest number of terrorist attacks.

I also analyzed the types of weapons used in terrorist attacks. I created a pie chart to visualize the distribution of weapon types. The chart showed that explosives are the most commonly used weapon in terrorist attacks, followed by firearms and incendiary devices.

I further explored the data by looking at the distribution of terrorist attacks by target type. I created a bar chart to visualize the top 10 target types. The chart showed that private citizens and property are the most common targets of terrorist attacks, followed by military and police targets.

Finally, I examined the relationship between the number of fatalities and injuries in terrorist attacks. I created a scatter plot using the Seaborn library to visualize the relationship. The plot showed that there is a positive correlation between the number of fatalities and injuries in terrorist attacks.

The analysis of the GTD revealed several patterns and trends in terrorism over the years. The dataset showed a sharp increase in the number of terrorist attacks after 2000, with Iraq, Pakistan, and Afghanistan being the countries most affected by terrorism. Explosives are the most commonly used weapon, and private citizens and property are the most common targets of terrorist attacks. I also found that there is a positive correlation between the number of fatalities and injuries in terrorist attacks.

The insights from this project can inform policy decisions and guide counter-terrorism efforts. In conclusion, the EDA of the Global Terrorism Database provided valuable insights into the dynamics of terrorism worldwide. The dataset is a rich source of information that can be used to further analyze the patterns and trends in terrorism and inform policy decisions.

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
df=df[["eventid","iyear","imonth","iday","country_txt","region_txt", "provstate", "city","latitude","longitude","location","targtype1_txt", "attacktype1_txt", "weaptype1_txt","gname","motive","nkill","nwound"]]
````

````python
# renamimg the column name to the relevant names
df.rename(columns={"iyear":"year","imonth":"month", "iday":"day", "country_txt":"country","region_txt":"region", "provstate":"state","targtype1_txt":"target","attacktype1_txt":"attack","weaptype1_txt":"weapon","gname":"groupname","nkill":"killed","nwound":"wounded"},inplace=True)
````

````python
# filling the null values of the killed and the wounded columns by 0
df["killed"]=df["killed"].fillna(0)
df["wounded"]=df["wounded"].fillna(0)
````

````python
# Defining the casualties columns into the dataframe 
df["casualties"]=df["killed"]+df["wounded"]
````

````python
#replacing the NaNs with "empty"
for i in df:
  df[i]=df[i].replace(np.nan , "empty")
df.head()
````

## Analysing the country wise trend of terrorist attacks

In general, a correlation heatmap chart is used to visualize the correlation between pairs of variables in a dataset. It is a useful tool to quickly identify patterns and relationships between variables, and it can be particularly helpful in exploratory data analysis.

 A correlation heatmap allows you to easily spot which variables are strongly correlated (either positively or negatively) and which ones are not. 

 Additionally, the correlation heatmap can also help us detect potential multicollinearity issues, which occur when two or more variables are highly correlated with each other, and therefore provide redundant information.
